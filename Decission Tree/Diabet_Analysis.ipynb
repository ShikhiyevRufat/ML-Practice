```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings

warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_diabetes

data = load_diabetes()

data
```

```python
print(data['DESCR'])
```

```python
X = pd.DataFrame(data['data'], columns=['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
X
```

```python
y = data['target']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=42)
```

```python
X_train.corr()
```

```python
import seaborn as sns

plt.figure(figsize=[15, 10])
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor



param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}

tree_regressor = DecisionTreeRegressor()
grid = GridSearchCV(tree_regressor, param, cv = 5, scoring= "neg_mean_squared_error")

grid

```

```python
grid.fit(X_train, y_train)

y_pred = grid.predict(X_test)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
selectedmodel=DecisionTreeRegressor(criterion='squared_error',max_depth=2,max_features='sqrt',splitter='best')
```

```python
selectedmodel.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectedmodel,filled=True)
```
