```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Travel.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop('CustomerID', axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Gender'].value_counts()
```

```python
df['MaritalStatus'].value_counts()
```

```python
df['Gender'] = df['Gender'].replace('Fe Male', 'Female')
df['MaritalStatus'] = df['MaritalStatus'].replace('Unmarried', 'Single')
```

```python
df['MaritalStatus'].value_counts()
```

```python
columns_na = [col for col in df.columns if df[col].isnull().sum() >= 1]

for columns in columns_na:
    print(columns, np.round(df[columns].isnull().mean()*100, 5), '%missing values')
```

```python
df.Age.fillna(df.Age.median(), inplace=True)

df.TypeofContact.fillna(df.TypeofContact.mode()[0], inplace=True)

df.DurationOfPitch.fillna(df.DurationOfPitch.median(), inplace=True)

df.NumberOfFollowups.fillna(df.NumberOfFollowups.mode()[0], inplace=True)

df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode()[0], inplace=True)

df.NumberOfTrips.fillna(df.NumberOfTrips.median(), inplace=True)

df.NumberOfChildrenVisiting.fillna(df.NumberOfChildrenVisiting.mode()[0], inplace=True)

df.MonthlyIncome.fillna(df.MonthlyIncome.median(), inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['TotalVisiting'] = df['NumberOfPersonVisiting'] + df['NumberOfChildrenVisiting']
df.drop(columns=['NumberOfChildrenVisiting', 'NumberOfPersonVisiting'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('ProdTaken', axis=1)
y = df['ProdTaken']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

X_train.head()
```

```python
numeric_column = X.select_dtypes(exclude='object').columns
categoric_column = X.select_dtypes(include='object').columns
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

numeric_transform = StandardScaler()
categoric_transform = OneHotEncoder(drop='first')

preprocessing = ColumnTransformer(
    [
        ("OneHotEncoder", categoric_transform, categoric_column),
        ("Numeric Transform", numeric_transform, numeric_column)
    ]
)

preprocessing
```

```python
X_train = preprocessing.fit_transform(X_train)
X_test=preprocessing.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
X_test
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, \
                            precision_score, recall_score, f1_score, roc_auc_score,roc_curve
```

```python
models = {
    'LinearRegression': LogisticRegression(),
    'RandomForestClassifier': RandomForestClassifier(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'GradientBoostingClassifier': GradientBoostingClassifier()
}


for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_train)

    y_train_predict = model.predict(X_train)
    y_test_predict = model.predict(X_test)



    # Train
    model_train_accuracy = accuracy_score(y_train, y_train_predict) 
    model_train_f1 = f1_score(y_train, y_train_predict, average='weighted') 
    model_train_precision = precision_score(y_train, y_train_predict) 
    model_train_recall = recall_score(y_train, y_train_predict) 
    model_train_rocauc_score = roc_auc_score(y_train, y_train_predict)
    # Test
    model_test_accuracy = accuracy_score(y_test, y_test_predict) 
    model_test_f1 = f1_score(y_test, y_test_predict, average='weighted') 
    model_test_precision = precision_score(y_test, y_test_predict) 
    model_test_recall = recall_score(y_test, y_test_predict) 
    model_test_rocauc_score = roc_auc_score(y_test, y_test_predict)


    print(f"Your model is: {list(models.values())[i]}")

    print("Train performances: ")
    print('Model performance for Training set')
    print("- Accuracy: {:.4f}".format(model_train_accuracy))
    print('- F1 score: {:.4f}'.format(model_train_f1))
    print('- Precision: {:.4f}'.format(model_train_precision))
    print('- Recall: {:.4f}'.format(model_train_recall))
    print('- Roc Auc Score: {:.4f}'.format(model_train_rocauc_score))

    print("-"*50)

    print('Test performances: ')
    print('- Accuracy: {:.4f}'.format(model_test_accuracy))
    print('- F1 score: {:.4f}'.format(model_test_f1))
    print('- Precision: {:.4f}'.format(model_test_precision))
    print('- Recall: {:.4f}'.format(model_test_recall))
    print('- Roc Auc Score: {:.4f}'.format(model_test_rocauc_score))

    print("*"*50)
    print('\n')
```

```python
rf_params = {"max_depth": [5, 8, 15, None, 10],
             "max_features": [5, 7, "auto", 8],
             "min_samples_split": [2, 8, 15, 20],
             "n_estimators": [100, 200, 500, 1000]}
```

```python
randomcv_models = [
                   ("RF", RandomForestClassifier(), rf_params)
                   ]
```

```python
from sklearn.model_selection import RandomizedSearchCV

model_param = {}

for name, model, param in randomcv_models:
    rsv = RandomizedSearchCV(estimator=model, param_distributions=param, cv=5, verbose=2)
    rsv.fit(X_train, y_train)
    model_param[name] = rsv.best_params_

for model in model_param:
    print(f"---------------- Best Params for {model} -------------------")
    print(model_param[model])

```
